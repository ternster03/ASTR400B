```python
class M33AnalyticOrbit:
    """ Calculate the analytical orbit of M33 around M31 """
    def __init__(self, output_file):
         """
        Initializes the class, computes the initial position and velocity vectors for M33 and M31.
        
        Parameters:
        output_file : str
            The name of the output file where the orbital results will be saved.
        
        Sets:
        self.r0 : numpy array
            The initial relative position vector of M33 with respect to M31.
        self.v0 : numpy array
            The initial relative velocity vector of M33 with respect to M31.
        """
        # Gravitational constant in appropriate units
        self.G = const.G.to(u.kpc**3/u.Msun/u.Gyr**2).value
        
        # Store the output file name
        self.output_file = output_file
        
        # Create an instance of the CenterOfMass class for M33 and M31 to get their COM positions and velocities
        M33_COM = CenterOfMass("M33_000.txt", 2)
        posM33 = M33_COM.COM_P(0.1)  # Position of M33's COM
        vosM33 = M33_COM.COM_V(posM33[0],posM33[1],posM33[2])  # Velocity of M33's COM

        M31_COM = CenterOfMass("M31_000.txt", 2)
        posM31 = M31_COM.COM_P(0.1)  # Position of M31's COM
        vosM31 = M31_COM.COM_V(posM31[0],posM31[1],posM31[2]) # Velocity of M31's COM

        # Store the initial relative position and velocity (M33 - M31)
        self.r0 = posM33.value - posM31.value
        self.v0 = vosM33.value - vosM31.value
        
        ### store the DIFFERENCE between the vectors posM33 - posM31
        # **** create two VECTORs self.r0 and self.v0 and have them be the
        # relativ
        ### get the mass of each component in M31 
        ### disk
        # **** self.rdisk = scale length (no units)
        self.rdisk = 5
        # **** self.Mdisk set with ComponentMass function. Remember to *1e12 to get the right units. Use the right ptype
        self.Mdisk = .12 * 10e12
        ### bulge
        # **** self.rbulge = set scale length (no units)
        self.rbulge = 1
        # **** self.Mbulge  set with ComponentMass function. Remember to *1e12 to get the right units Use the right ptype
        self.Mbulge = .019 * 10e12
        # Halo
        # **** self.rhalo = set scale length from HW5 (no units)
        self.rhalo = 60 
        # **** self.Mhalo set with ComponentMass function. Remember to *1e12 to get the right units. Use the right ptypee position and velocity VECTORS of M33
        self.Mhalo = 1.921 * 10e12
        
     
    
    
    def HernquistAccel(self,M, r_a, r): # it is easiest if you take as an input the position VECTOR 
        """
        Compute the acceleration due to the Hernquist potential at a given position.

        Parameters:
        M : float
            The mass of the object (e.g., Mhalo for the halo mass).
        r_a : float
            A characteristic scale length of the Hernquist profile.
        r : numpy array
            The position vector of the object relative to the center of M31.

        Returns:
        Hern : numpy array
            The acceleration vector at position r.
        """
        
        ### Store the magnitude of the position vector
        rmag = np.linalg.norm(r)
        # Prob wrong
        ### Store the Acceleration
        Hern =  (-self.G*M)/(rmag *(r_a + rmag)**2) * r
        # NOTE: we want an acceleration VECTOR so you need to make sure that in the Hernquist equation you 
        # use  -G*M/(rmag *(ra + rmag)**2) * r --> where the last r is a VECTOR 
        
        return Hern
    
    
    
    def MiyamotoNagaiAccel(self, M, r_d, r):# it is easiest if you take as an input a position VECTOR  r 
       """
        Compute the acceleration due to the Miyamoto-Nagai potential at a given position.

        Parameters:
        M : float
            The mass of the object (e.g., Mdisk for the disk mass).
        r_d : float
            A characteristic scale length of the Miyamoto-Nagai profile (disk scale).
        r : numpy array
            The position vector of the object relative to the center of M31.

        Returns:
        Miya : numpy array
            The acceleration vector at position r.
        """
        
        ### Acceleration **** follow the formula in the HW instructions
        
        # AGAIN note that we want a VECTOR to be returned  (see Hernquist instructions)
        # this can be tricky given that the z component is different than in the x or y directions. 
        # we can deal with this by multiplying the whole thing by an extra array that accounts for the 
        # differences in the z direction:
        #  multiply the whle thing by :   np.array([1,1,ZSTUFF]) 
        # where ZSTUFF are the terms associated with the z direction
        R = np.sqrt(r[0]**2 + r[1]**2)
        z_d = r_d/5
        B = r_d + np.sqrt(r[2]**2 + z_d)
        
        Miya = -self.G*M/((R**2 +B**2)**(1.5)) * r
        
        Miya = np.array([1,1,B/np.sqrt(r[2]**2+z_d**2)])
        
        return Miya
        # the np.array allows for a different value for the z component of the acceleration
     
    
    def M31Accel(self, r): # input should include the position vector, r
        """
        Compute the total acceleration on M33 due to the gravity of 
        the halo, bulge, and disk components of M31.

        Parameters:
        r : numpy array
            The position vector of M33 relative to M31.

        Returns:
        tot_acc : numpy array
            The total acceleration vector on M33.
        """
        ### Call the previous functions for the halo, bulge and disk
        disk = self.MiyamotoNagaiAccel(self.Mdisk,self.rdisk,r)
        halo = self.HernquistAccel(self.Mhalo, self.rhalo, r)
        bulge = self.HernquistAccel(self.Mbulge, self.rbulge, r)
        # these functions will take as inputs variable we defined in the initialization of the class like 
        # self.rdisk etc. 
        tot_acc = disk + halo + bulge    
            # return the SUM of the output of the acceleration functions - this will return a VECTOR 
        return tot_acc
    
    
    
    def LeapFrog(self,dt,r,v): # take as input r and v, which are VECTORS. Assume it is ONE vector at a time
        """
        Perform one time step of the Leap Frog integration to update the position and velocity of M33.

        Parameters:
        dt : float
            The time step for the integration.
        r : numpy array
            The current position vector of M33 relative to M31.
        v : numpy array
            The current velocity vector of M33 relative to M31.

        Returns:
        rnew : numpy array
            The updated position vector of M33.
        vnew : numpy array
            The updated velocity vector of M33.
        """
        # predict the position at the next half timestep
        rhalf = r + v * dt/2
        
        a_half = self.M31Accel(rhalf)
        # predict the final velocity at the next timestep using the acceleration field at the rhalf position 
        vnew = v + a_half * dt
        
        # predict the final position using the average of the current velocity and the final velocity
        # this accounts for the fact that we don't know how the speed changes from the current timestep to the 
        # next, so we approximate it using the average expected speed over the time interval dt. 
        rnew = rhalf + vnew * dt/2
        
        return rnew, vnew
    
    
    
    def OrbitIntegration(self, t0, dt, tmax):
        """
        Integrate the orbit of M33 using the Leap Frog method over the specified time range.

        Parameters:
        t0 : float
            The starting time of the integration.
        dt : float
            The time step for the integration.
        tmax : float
            The final time of the integration.

        Returns:
        Results: Results from function
        """
        # initialize the time to the input starting time
        t = t0
        
        # initialize an empty array of size :  rows int(tmax/dt)+2  , columns 7
        orbit = np.zeros([int(tmax/dt + 2), 7])
        
        # initialize the first row of the orbit
        orbit[0] = t0, *tuple(self.r0), *tuple(self.v0)
        # this above is equivalent to 
        # orbit[0] = t0, self.r0[0], self.r0[1], self.r0[2], self.v0[0], self.v0[1], self.v0[2]
        self.r = self.r0
        self.v = self.v0
        # initialize a counter for the orbit.  
        i = 1 # since we already set the 0th values, we start the counter at 1
        
        # start the integration (advancing in time steps and computing LeapFrog at each step)
        while (t <= tmax):  # as long as t has not exceeded the maximal time 
            
            # **** advance the time by one timestep, dt
            t = t + dt
            # **** store the new time in the first column of the ith row
            orbit[i][0] = t
            
            # ***** advance the position and velocity using the LeapFrog scheme
            # remember that LeapFrog returns a position vector and a velocity vector  
            # as an example, if a function returns three vectors you would call the function and store 
            # the variable like:     a,b,c = function(input)
            self.r, self.v = self.LeapFrog(dt,self.r,self.v)
         
    
            # ****  store the new position vector into the columns with indexes 1,2,3 of the ith row of orbit
            # TIP:  if you want columns 5-7 of the Nth row of an array called A, you would write : 
            # A[n, 5:8] 
            # where the syntax is row n, start at column 5 and end BEFORE column 8
            
            # ****  store the new position vector into the columns with indexes 1,2,3 of the ith row of orbit
            orbit[i,1:4] = self.r
            orbit[i,4:7] = self.v
            # **** update counter i , where i is keeping track of the number of rows (i.e. the number of time steps)
            i +=1
        
        
        # write the data to a file
        np.savetxt(self.output_file, orbit, fmt = "%11.3f"*7, comments='#', 
                   header="{:>10s}{:>11s}{:>11s}{:>11s}{:>11s}{:>11s}{:>11s}"\
                   .format('t', 'x', 'y', 'z', 'vx', 'vy', 'vz'))
        
        # there is no return function
        
        
```

```python
# function to compute the magnitude of the difference between two vectors 
# You can use this function to return both the relative position and relative velocity for two 
# galaxies over the entire orbit  
def relative_mag(a, b): 
    """
    Function that computes the magnitude of the difference between two vectors.
    Inputs with shape (3, n) will return n outputs

    PARAMETERS
    ----------
    a : `np.ndarray'
        first vector
    b : 'np.ndarray'
        second vector

    RETURNS
    -------
    mag : `float or np.ndarray`
        |a-b|
    """
    
    # compute the difference vector
    x = a[0] - b[0] 
    y = a[1] - b[1]
    z = a[2] - b[2]

    # return its magnitude
    return np.sqrt(x**2 + y**2 + z**2)
```

```python

# position vectors of each galaxy with respect to 0,0,0
M31_pos = np.array([M31_orbit['x'], M31_orbit['y'], M31_orbit['z']])
M33_pos = np.array([M33_orbit['x'], M33_orbit['y'], M33_orbit['z']])
M31_M33_pos = ((Homework7_output['x'])**2 + (Homework7_output['y'])**2 + (Homework7_output['z'])**2)**(1/2)
# velocity vectors of each galaxy with respect to 0,0,0
M31_vel = np.array([M31_orbit['vx'], M31_orbit['vy'], M31_orbit['vz']])
M33_vel = np.array([M33_orbit['vx'], M33_orbit['vy'], M33_orbit['vz']])
M31_M33_vos = ((Homework7_output['vx'])**2 + (Homework7_output['vy'])**2 + (Homework7_output['vz'])**2)**(1/2)
```

```python
# Determine the magnitude of the relative position and velocities 

# of M33 and M31
M33_M31_relpos = relative_mag(M31_pos, M33_pos)
M33_M31_relvel = relative_mag(M31_vel, M33_vel)
```
