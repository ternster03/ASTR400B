# Part A

Create a function that returns the SFR for a given luminosity (NUV, FUV, TIR, Halpha)

$Log( {\rm SFR} (M_\odot/year)) = Log(Lx (erg/s)) - Log(Cx)$ 

Including corrections for dust absorption 

Kennicutt & Evans 2012 ARA&A Equation 12 and Table 1, 2

```python
def StarFormationRate(L,Type, TIR = 0):
    """
    Function that computes the star formation rate of a galaxy.
    This follows the work in Kennicutt and Evans 2012 Eq 12 (AR&A 50)
    
    Inputs: L (float): Lumniosity of the galaxy in ergs/s
    Type (str): Indicates the wavelength: "FUV", "NUV", "TIR", "Halpha"
    TIR (float): Total Infrared Luminosity in ergs/s (default = 0)
    
    
    Outputs: SFT(float) Log of the Star Formation Rate (Msun/yr)
    """
    if Type == 'FUV':
        logCx = 43.35
        TIRc = .46
    elif Type == 'NUV':
        logCx = 43.17
        TIRc = .27
    elif Type == 'Halpha':
        logCx = 41.27
        TIRc = .0024
    elif Type == 'TIR':
        logCx = 43.41
        TIRc = 0
    else:
        print(f"Type '{Type}' is not a valid type. Should be one of 'FUV', 'NUV', 'TIR', or 'Halpha'")
        return
    # Correct the luminosity fo rdust using the TIR
    Lcorr = L + TIRc*TIR
    # Star formation rate
    SFR = np.log10(Lcorr)- logCx
    return SFR
```

# Part B Star formation main sequence

1) Write a function that returns the average SFR of a galaxy at a given redshift, given its stellar mass

2) What is the average SFR of a MW mass galaxy today? at z=1?

3) Plot the SFR main sequence for a few different redshifts from 1e9 to 1e12 Msun.


From Whitaker 2012:

log(SFR) = $\alpha(z)({\rm log}M_\ast - 10.5) + \beta(z)$

$\alpha(z) = 0.7 - 0.13z$

$\beta(z) = 0.38 + 1.14z - 0.19z^2$

# Step 1

```python
def SFRMainSequence(Mstar,z):
    """
    Function that computes the average star formation rate of a galaxy.
    As a funciton of stellar mass and redshift.
    Inputs:
    Mstar (float): Stellar mass of the galaxy in Msun
    z (float): redshift of the galaxy
    
    Outputs: SFR (float):
    log of the SFR log(Msun/yr)
    """
    alpha = 0.7 -.13*z
    beta = .38 + 1.14*z -.19*z*z
    SFR = alpha*(np.log10(Mstar)-10.5) + beta
    return SFR
```

# Step 2

```python
# MW at z=0
MWmstar = 7.5e10
```

```python
print(SFRMainSequence(MWmstar,0))
print(10**SFRMainSequence(MWmstar,0))
```

```python
# MW at z = 1
print(SFRMainSequence(MWmstar,1))
print(10**SFRMainSequence(MWmstar,1))
```

# Step 3

```python
# create an array of stellar masses
masses = np.linspace(1e8,1e12)
```

```python

fig = plt.figure(figsize=(8,8), dpi=500)
ax = plt.subplot(111)

# add log log plots
ax.plot(np.log10(masses),SFRMainSequence(masses,0), color = "red", linestyle = '-', label = "z=0")
ax.plot(np.log10(masses),SFRMainSequence(masses,1), color = "blue", linestyle = '-.', label = "z=1")
ax.plot(np.log10(masses),SFRMainSequence(masses,2), color = "black", linestyle = ':', label = "z=2")
ax.plot(np.log10(masses),SFRMainSequence(masses,3), color = "green", linestyle = '--', label = "z=3")
# Add axis labels
plt.xlabel('Log(Mstar (M$_\odot$))', fontsize=12)
plt.ylabel('Log(SFR (M$_\odot$/year))', fontsize=12)


#adjust tick label font size
label_size = 12
matplotlib.rcParams['xtick.labelsize'] = label_size 
matplotlib.rcParams['ytick.labelsize'] = label_size

# add a legend with some customizations.
legend = ax.legend(loc='upper left',fontsize='x-large')


# Save file
#plt.savefig('Lab8_SFR_MainSequence.png')
```
